# World Happiness

A partir de un encuesta a una muestra significativa de la población de un país, se ha obtenido un marcador de felicidad. Los otros datos son proporcionados por el govierno de cada país.

Se tiene las información de los años 2015, 2016 y 2017.

Los datos tienen las siguienes columnas:
- **Country:** País
- **Happiness Rank:** Puesto que ocupa el país según su felicidad
- **Happiness Score:** Marcador de felicidad
- **Economy (GDP per Capita):** Economía del país (PIB)
- **Family:** Familia
- **Health (Life Expectancy):** Salud (esperanza de vida)
- **Freedom:** Libertad
- **Trust (Government Corruption):** Confianza en el govierno (según corrupción)
- **Generosity:** Genorisidad
- **Dystopia Residual:** Caso de menor felicidad del país

# Training a Linear Regression Model

### X and y arrays

```python
X = WH[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 
                    'Freedom', 'Generosity', 'Trust (Government Corruption)']]

y= WH['Happiness Score']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=87)
```

## Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

## Model Evaluation

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

## Predictions from our Model

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

#### Resiudual histogram

```python
sns.distplot((y_test-predictions),bins=40);
```

## Regression Evaluation Metrics

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
